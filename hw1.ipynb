```python
# Function to match spectrum m/z values with peptide masses (including modifications)
# tolerance is used here to account for the standards
def match_spectrum_with_modifications(spectrum, peptides, tolerance=0.2):
    annotations = []

    # Iterate over each m/z value in the spectrum
    # iterate over each row in the spectrum df and extracts the m/z values to find a match 
    for index, spectrum_row in spectrum.iterrows():
        mz = spectrum_row['m/z']
        peptide_match = None

        # Iterate over each peptide to find a match
        # loop will iterate over each row in df and check the differences base on the tolerance set above 
        # loop will break and start over once a match is found 
        for _, peptide_row in peptides.iterrows():
            # Check both original and modified masses for a match within the tolerance range
            if abs(peptide_row['mass'] - mz) <= tolerance or abs(peptide_row['modified_mass'] - mz) <= tolerance:
                peptide_match = peptide_row['peptide_sequence']
                break

        if peptide_match:
            annotations.append((mz, peptide_match))

    return annotations

# Perform the matching
matched_annotations = match_spectrum_with_modifications(spectrum_df, peptides_df_with_modifications)

# Convert to DataFrame for better visualization
matched_annotations_df = pd.DataFrame(matched_annotations, columns=['m/z', 'Peptide Sequence'])

# Display the DataFrame with matched annotations
print(matched_annotations_df)
```
